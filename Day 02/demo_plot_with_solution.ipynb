## Demos

### scatter() and plot()
- What is the difference between the two functions?

```python
x1 = [0,3,5,4]
y1 = [1,3,2,1]
plt.scatter(x1,y1)
plt.xlabel('time')
plt.ylabel('score')
plt.grid()
```

```python
plt.plot(x1,y1)
plt.xlabel('x')
plt.ylabel('y')
```

### Straight line
- plot() connects points with a line
- to plot f(x), we need to generate a lot of points (x, f(x))
- use linspace() to generate x coordiantes

```python
x = np.linspace(-1,5,100)
print(x)
```

- is y2 a function, a scalar or a vector? 
- what is each element of y2?

```python
y2 = 2*x + 0.5
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

```python
## All-in-one plot
plt.figure()  #figsize = (8,8)

plt.plot(x1, y1, 'o')
plt.plot(x, y2)  # c = 'green'  change color
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)

## After add scatter points, need to change x to (-1,5) to get better plot
plt.ylim([-2, 4])
plt.xlim([-1, 5])
plt.xlabel('x')
plt.ylabel('y')
plt.title('All-in-one Plot')
plt.grid()

plt.legend(['Linear','Sine','Poly','Expo','Scatter'])
```
